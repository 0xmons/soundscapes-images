```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dim = 300
directions = [[0,1], [0,-1], [-1,0], [1,0]]
diagonal_directions = [[1,1], [1,-1], [-1,-1], [1,-1]]

def random_walk(arr, length, x, y):
  prev_x = x
  prev_y = y
  for i in range(length):
    rng =int(np.random.random()*4)
    dx = directions[rng][0]
    dy = directions[rng][1]
    prev_x = x
    prev_y = y
    x = x + dx
    y = y + dy
    if (x == prev_x and y == prev_y):
      rng =int(np.random.random()*4)
      dx = diagonal_directions[rng][0]
      dy = diagonal_directions[rng][1]
      prev_x = x
      prev_y = y
      x = x + dx
      y = y + dy
    try:
      arr[x][y] = 255
    except IndexError:
      arr[int(dim/2)][int(dim/2)] = 255
```

```python
for k in range(341):  
  maze = np.zeros((dim,dim))
  for i in range(10):
    random_walk(maze, 10000, int(dim/(i+1)), int(dim/(i+1)))
#   plt.imshow(maze,cmap='gray', vmin = 0, vmax = 255)
  plt.imsave('./mazes/' + str(k) + '.png', maze, cmap='gray')
```
